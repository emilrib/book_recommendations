```python
import pandas as pd
```

### Process the Genres

```python
# Read books table
df = pd.read_csv("C:/Users/Christina/OneDrive - Hochschule Luzern/DBM Project/Data/books_1.Best_Books_Ever.csv")
```

```python
# Delete columns not needed for the genre table
genre = df.drop(["bookId", "title", "series", "author", "rating", "description", "language", "characters", "bookFormat", "edition", "pages", "publisher", "publishDate", "firstPublishDate", "awards", "numRatings", "ratingsByStars", "likedPercent", "setting", "coverImg", "bbeScore", "bbeVotes", "price"], axis=1)
```

```python
# Drop all books without valid isbn
genre.drop(genre[genre['isbn'] == "9999999999999"].index, inplace = True)
```

```python
# Split the rows with repeating groups into separate rows
genre = genre.assign(genres=genre.genres.str.split(',')).explode('genres')
```

```python
# Save new table into csv
genre.to_csv("C:/Users/Christina/OneDrive - Hochschule Luzern/DBM Project/Data/genres.csv", index=False)
```

Note: further cleaning took place in Excel

```python
# Upload cleaned genre table
genre2 = pd.read_csv("C:/Users/Christina/OneDrive - Hochschule Luzern/DBM Project/Data/genres_new.csv", sep=";")
```

```python
# Get amount of values
genre2
```

```python
# Amount of unique values
genre2['genre'].nunique()
```

There are 382186 rows with 981 unique values in the genre table

```python
# Separate Genres into new table
genre_id = pd.DataFrame(genre2['genre'].unique(), columns=["genre"])
```

```python
# Add unique ID to genres
genre_id.insert(0, 'Genre_ID', range(1, 1 + len(genre_id)))
genre_id
```

```python
# Replace genre with genre ID in genre2
genre2['genre'] = genre2['genre'].map(genre_id.set_index('genre')['Genre_ID'])
genre2
```

```python
# Rename the column
genre2.rename(columns = {'genre':'genre_id'}, inplace = True)
genre2
```

```python
# Save both tables
genre2.to_csv("C:/Users/Christina/OneDrive - Hochschule Luzern/DBM Project/Data/book_genres_final.csv", index=False)
genre_id.to_csv("C:/Users/Christina/OneDrive - Hochschule Luzern/DBM Project/Data/genres_ids.csv", index=False)
```
